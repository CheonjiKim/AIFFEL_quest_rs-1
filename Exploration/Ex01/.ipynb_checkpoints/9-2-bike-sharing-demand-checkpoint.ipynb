```markdown
# 라이브러리 임포트

필요한 라이브러리를 불러옵니다.
```


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import minmax_scale
import calendar
from datetime import datetime
````

```markdown
# (1) 데이터 가져오기

데이터셋을 로드합니다. 경로는 환경에 따라 변경될 수 있습니다.
```

````python
train = pd.read_csv('~/data/data/bike-sharing-demand/train.csv')
train.head()
````

```markdown
# (2) datetime 컬럼을 datetime 자료형으로 변환하고 연, 월, 일, 시, 분, 초까지 6가지 컬럼 생성하기

이미 datetime 형식이라면 변환하지 않아도 됩니다.
```

````python
train['datetime'] = pd.to_datetime(train['datetime'])
train['year']=train[ "datetime"].dt.year
train['month']=train[ "datetime"].dt.month
train['day']=train[ "datetime"].dt.day
train['hour']=train[ "datetime"].dt.hour
train['minute']=train[ "datetime"].dt.minute
train['second']=train[ "datetime"].dt.second
train.info()
````

```markdown
# (3) year, month, day, hour, minute, second 데이터 개수 시각화하기

서브플롯을 사용하여 한 번에 6개의 그래프를 시각화합니다.
```

````python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 6))
plt.subplots_adjust(wspace=0.3, hspace=0.4)
plt.suptitle('datetime counts', fontsize=20)
sns.countplot(x='year', data=train, ax=axes[0, 0])
sns.countplot(x='month', data=train, ax=axes[0, 1])
sns.countplot(x='day', data=train, ax=axes[1, 0])
sns.countplot(x='hour', data=train, ax=axes[1, 1])
sns.countplot(x='minute', data=train, ax=axes[2, 0])
sns.countplot(x='second', data=train, ax=axes[2, 1])

plt.show()
````

```markdown
# (4) 피쳐 선택 및 스케일링

불필요한 피처들을 제거하고, 나머지 피처들에 대해 MinMax 스케일링을 수행합니다.
```

````python
feature_cols = ['year', 'month', 'day', 'hour']
X = train[feature_cols]
y = train['count']

# MinMax 스케일링
X_scaled = minmax_scale(X)
````

```markdown
# (5) 학습 및 테스트 데이터 분할

`train_test_split` 함수로 학습 데이터와 테스트 데이터를 8:2의 비율로 분할합니다.
```

````python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
````

```markdown
# (6) 모델 학습 및 평가

선형 회귀 모델을 학습하고, 테스트 데이터에 대한 예측 결과를 이용하여 Mean Squared Error (MSE)와 R² 값을 출력합니다.
```

````python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R² Score:", r2)
````

```markdown
# (7) 결과 시각화

온도와 습도에 따른 실제값과 예측값을 산점도로 비교해봅니다.
```

````python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(train['temp'], y, color='blue', alpha=0.5, label='Actual')
plt.scatter(train['temp'], model.predict(X_scaled), color='red', alpha=0.5, label='Predicted')
plt.xlabel('Temperature')
plt.ylabel('Count')
plt.title('Actual vs Predicted Count by Temperature')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(train['humidity'], y, color='blue', alpha=0.5, label='Actual')
plt.scatter(train['humidity'], model.predict(X_scaled), color='red', alpha=0.5, label='Predicted')
plt.xlabel('Humidity')
plt.ylabel('Count')
plt.title('Actual vs Predicted Count by Humidity')
plt.legend()

plt.tight_layout()
plt.show()
````
